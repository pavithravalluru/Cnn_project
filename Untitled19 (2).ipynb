```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, BatchNormalization, ReLU, Input, concatenate, UpSampling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from PIL import Image
import matplotlib.pyplot as plt
from tensorflow.keras.applications import VGG16
from tensorflow.keras import backend as K
import tensorflow.image as tf_img

gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        tf.config.experimental.set_memory_growth(gpus[0], True)
    except RuntimeError as e:
        print(e)


def create_masked_images(images, mask_fraction=0.3):
    masked_images = images.copy()
    mask = np.random.rand(*images.shape[:3]) < mask_fraction
    for i in range(images.shape[0]):
        masked_images[i][mask[i]] = 0
    return masked_images


def load_images_from_folder(folder, img_size=(256, 256), max_images=500):
    def process_image(filename):
        img = Image.open(filename).convert('RGB')
        img = img.resize(img_size)
        return np.array(img) / 255.0

    images = [process_image(os.path.join(folder, f)) for f in os.listdir(folder) if f.lower().endswith(('png', 'jpg', 'jpeg', 'tif', 'bmp'))][:max_images]
    return np.array(images) if images else ValueError("No valid images found.")


CLEAN_IMG_FOLDER = "/content/drive/MyDrive/data/green_area"
clean_images = load_images_from_folder(CLEAN_IMG_FOLDER)


corrupted_images = create_masked_images(clean_images, mask_fraction=0.3)


x_train, x_val, y_train, y_val = train_test_split(corrupted_images, clean_images, test_size=0.2, random_state=42)


def create_tf_dataset(x, y, batch_size=4):
    dataset = tf.data.Dataset.from_tensor_slices((x, y))
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset

dataset_train = create_tf_dataset(x_train, y_train)
dataset_val = create_tf_dataset(x_val, y_val)

def build_inpainting_cnn(input_shape):
    input_layer = Input(shape=input_shape)
    x = Conv2D(32, (3, 3), padding='same', activation='relu')(input_layer)
    x = BatchNormalization()(x)
    for _ in range(2):
        x = Conv2D(32, (3, 3), padding='same', activation='relu')(x)
        x = BatchNormalization()(x)
    x = Conv2D(3, (3, 3), padding='same', activation='sigmoid')(x)
    return Model(inputs=input_layer, outputs=x)


vgg = VGG16(include_top=False, weights='imagenet', input_shape=(256, 256, 3))
vgg.trainable = False
vgg_partial = Model(inputs=vgg.input, outputs=vgg.layers[4].output)
def perceptual_loss(y_true, y_pred):
    y_true_resized = tf.image.resize(y_true, (256, 256))
    y_pred_resized = tf.image.resize(y_pred, (256, 256))
    y_true_features = vgg_partial(y_true_resized)
    y_pred_features = vgg_partial(y_pred_resized)
    return 0.1 * K.mean(K.square(y_true_features - y_pred_features))

def hybrid_loss(y_true, y_pred):
    perceptual = perceptual_loss(y_true, y_pred)
    mse = K.mean(K.square(y_true - y_pred))
    return perceptual + mse

def psnr_metric(y_true, y_pred):
    return tf_img.psnr(y_true, y_pred, max_val=1.0)

def ssim_metric(y_true, y_pred):
    return tf_img.ssim(tf.cast(y_true, tf.float32), y_pred, max_val=1.0)


optimizer = Adam(learning_rate=1e-4)
model = build_inpainting_cnn(input_shape=(256, 256, 3))
model.compile(optimizer=optimizer, loss=hybrid_loss, metrics=['mse', psnr_metric, ssim_metric])


history = model.fit(dataset_train, validation_data=dataset_val, epochs=20)


import matplotlib.pyplot as plt

def display_and_save_images(original, corrupted, reconstructed, output_folder="/content/drive/MyDrive/output_images"):
    os.makedirs(output_folder, exist_ok=True)

    num_images = len(original)


    for i in range(3):
        fig, axes = plt.subplots(1, 3, figsize=(12, 4))
        axes[0].imshow(original[i])
        axes[0].set_title("Original")
        axes[0].axis("off")

        axes[1].imshow(corrupted[i])
        axes[1].set_title("Corrupted")
        axes[1].axis("off")

        axes[2].imshow(reconstructed[i])
        axes[2].set_title("Reconstructed")
        axes[2].axis("off")

        plt.show()


    for i in range(num_images):
        Image.fromarray((original[i] * 255).astype(np.uint8)).save(f"{output_folder}/original_{i}.jpg")
        Image.fromarray((corrupted[i] * 255).astype(np.uint8)).save(f"{output_folder}/corrupted_{i}.jpg")
        Image.fromarray((reconstructed[i] * 255).astype(np.uint8)).save(f"{output_folder}/reconstructed_{i}.jpg")

    print(f"✅ Saved {num_images} images in {output_folder}")


reconstructed_images = model.predict(x_val, batch_size=4)


display_and_save_images(y_val, x_val, reconstructed_images)



plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss Over Epochs')
plt.show()


```

```python
import os
import numpy as np
import tensorflow as tf
import tensorflow.image as tf_img
import matplotlib.pyplot as plt
from PIL import Image

# Path to the folder containing all images
IMAGE_FOLDER = "/content/drive/MyDrive/output_images"

# Function to load and preprocess an image
def load_image(image_path, img_size=(256, 256)):
    img = Image.open(image_path).convert("RGB")
    img = img.resize(img_size)
    img_array = np.array(img) / 255.0  # Normalize to [0,1]
    return img_array

# Initialize lists to store error metrics
mse_errors = []
psnr_values = []
ssim_values = []

# Organize images by type using filename patterns
image_filenames = sorted(os.listdir(IMAGE_FOLDER))
original_images = [f for f in image_filenames if "original" in f.lower()]
corrupted_images = [f for f in image_filenames if "corrupted" in f.lower()]
reconstructed_images = [f for f in image_filenames if "reconstructed" in f.lower()]

# Ensure filenames match correctly
original_images.sort()
corrupted_images.sort()
reconstructed_images.sort()

# Process all images
for orig, corr, recon in zip(original_images, corrupted_images, reconstructed_images):
    orig_path = os.path.join(IMAGE_FOLDER, orig)
    corr_path = os.path.join(IMAGE_FOLDER, corr)
    recon_path = os.path.join(IMAGE_FOLDER, recon)

    # Load images
    original_image = load_image(orig_path)
    corrupted_image = load_image(corr_path)
    reconstructed_image = load_image(recon_path)

    # Convert to tensors for TensorFlow functions
    original_tensor = tf.convert_to_tensor(original_image, dtype=tf.float32)
    reconstructed_tensor = tf.convert_to_tensor(reconstructed_image, dtype=tf.float32)

    # Compute MSE
    mse = np.mean((original_image - reconstructed_image) ** 2)
    mse_errors.append(mse)

    # Compute PSNR
    psnr = -10 * np.log10(mse) if mse > 0 else 100  # Avoid log(0)
    psnr_values.append(psnr)

    # Compute SSIM
    ssim = tf_img.ssim(original_tensor, reconstructed_tensor, max_val=1.0).numpy()
    ssim_values.append(ssim)

# Plot histograms
plt.figure(figsize=(8, 5))
plt.hist(mse_errors, bins=10, color='red', alpha=0.7)
plt.xlabel("MSE Error")
plt.ylabel("Frequency")
plt.title("MSE Distribution of Reconstructed Images")
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(psnr_values, bins=10, color='green', alpha=0.7)
plt.xlabel("PSNR (dB)")
plt.ylabel("Frequency")
plt.title("PSNR Distribution of Reconstructed Images")
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(ssim_values, bins=10, color='blue', alpha=0.7)
plt.xlabel("SSIM")
plt.ylabel("Frequency")
plt.title("SSIM Distribution of Reconstructed Images")
plt.show()

# Print summary statistics
print(f"Average MSE: {np.mean(mse_errors):.6f}")
print(f"Average PSNR: {np.mean(psnr_values):.2f} dB")
print(f"Average SSIM: {np.mean(ssim_values):.4f}")
```
